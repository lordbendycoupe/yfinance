```python
# The function provides additional metrics for the moving averages
# - Difference
# -> +ve implies the short window moving average is above the long moving average (long position)
# -> -ve implies the short window moving average is below the long moving average (short position)
# - Converging or Diverging (in regards to the two SMAs) 
# -> C - Converging 
# -> D - Diverging

def conv_div(df, short_window, long_window):
    df2 = pd.DataFrame()
    df2['Close'] = df['Close']
    df2[f'{short_window}-day SMA'] = df[f'{short_window}-day SMA']
    df2[f'{long_window}-day SMA'] = df[f'{long_window}-day SMA']
    df2['Signal'] = df['Signal']
    df2['Difference b/w MAs'] = df2[f'{short_window}-day SMA']-df2[f'{long_window}-day SMA']
    diff = df2['Difference b/w MAs'].to_list()
    temp = [np.nan]*len(df2)
    for i in range(26, len(df2)):
        if abs(diff[i-1]) > abs(diff[i]):
            temp[i] = 'C'
        else:
            temp[i] = 'D'
    df2['Converging or Diverging (MA)'] = temp
    return df2
```

```python
# Measure the convergence and the divergence of share price from moving average.
def convergence(og):
    df = pd.DataFrame()
    df['Close'] = og['Close']
    df['9-day EMA'] = og['9-day EMA']
    df['26-day EMA'] = og['26-day EMA']
    df['Delta 26-Day EMA'] = df['Close'] - df["26-day EMA"]
    df['Delta 9-Day EMA'] = df['Close'] - df["9-day EMA"]
    return df
```
